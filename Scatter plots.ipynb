```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris') # Load Iris dataset
iris.head()
```

```python
sl = iris["sepal_length"]
sw = iris["sepal_width"]
```

```python
plt.plot(sl, sw, 'o')#This line of code uses the plot function from Matplotlib.
```

```python
plt.scatter(sl, sw)  # This line of code uses the scatter function from Matplotlib.
```

```python
plt.figure(figsize=(8, 4))# Set the size of the figure to be 8 inches in width and 4 inches in height
# Create a scatter plot using data from 'sl' (sepal length) and 'sw' (sepal width)
# 'marker='s'' specifies that square markers ('s') should be used
# 'color='red'' specifies that the markers should be colored red
plt.scatter(sl, sw, marker='s', color='black')
plt.xlabel('Sepal Length')# Label the x-axis as 'Sepal Length'
plt.ylabel('Sepal Width')# Label the y-axis as 'Sepal Width'
plt.title('Sepal Length vs Sepal Width') # Add a title to the plot
plt.show() # Display the plot
```

### Analyzing the Scatter Plot


```python
# Example 1: Linear Correlation example (Sepal Length vs Petal Length) demonstrates a linear correlation.
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Linear Correlation: Sepal Length vs Petal Length')
plt.show()
```

```python
# Example 2: Nonlinear Correlation example (Sepal Length vs Petal Length) demonstrates a linear correlation.
plt.scatter(iris['petal_width'], iris['sepal_width'])
plt.xlabel('Petal Width')
plt.ylabel('Sepal Width')
plt.title('Nonlinear Correlation: Petal Width vs Sepal Width')
plt.show()
```

```python
# Example 3: Strong Positive Correlation example (Sepal Length vs Petal Length) shows a strong positive correlation.
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Strong Positive Correlation: Sepal Length vs Petal Length')
plt.show()
```

```python
# Example 4: Weak Negative Correlation example (Sepal Length vs Sepal Width) exhibits a weak negative correlation.
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Weak Negative Correlation: Sepal Length vs Sepal Width')
plt.show()
```

```python
# Example 5: Negative Correlation  example (Petal Length vs Petal Width) displays a negative correlation.
plt.scatter(iris['petal_length'], iris['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Negative Correlation: Petal Length vs Petal Width')
plt.show()
```
